```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Read data from CSV files for feeding and control Bacteroides
df_F = pd.read_csv('taxa_abund_across_gt_p8_p28.csv', index_col=0).T
df_F
df_F['Lactobacillus'] = pd.to_numeric(df_F['Lactobacillus'], errors='coerce')
df_F['Escherichia_Shigella'] = pd.to_numeric(df_F['Escherichia_Shigella'], errors='coerce')
df_F['Bacteroides'] = pd.to_numeric(df_F['Bacteroides'], errors='coerce')
df_concat = df_F

# Actinobacteria
plt.figure(figsize=(10, 9))
ax = plt.axes()
# Set specific colors for DOL
colors = {
    'P8': '#F8766D',
    'P11': '#7CAE00',  
          'P14': '#00BFC4',
          'P28': '#C77CFF'}
dot_colors = {
    'P8': 'black',
    'P11': 'black',  
          'P14': 'black',
          'P28': 'black'}
dol_order = ['P8','P11', 'P14', 'P28']
# Plot data from df_concat
sns.boxplot(x='DOL', y='Lactobacillus', hue='DOL', data=df_concat, ax=ax, palette=colors, width=0.5, sym='', order=dol_order, dodge=False)

# Add swarm plot on top of boxplot inside each box group
sns.swarmplot(x='DOL', y='Lactobacillus', hue='DOL', data=df_concat, palette=dot_colors, dodge=False, order=dol_order, hue_order=dol_order,size=7,legend=False)

# Set labels and title
plt.title('Lactobacillus',fontsize=18)
plt.xlabel('DOL',fontsize=18)
plt.ylabel('Abundance',fontsize=18)

# Show the legend
plt.legend(title='DOL')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
# Actinobacteria
plt.figure(figsize=(10, 9))
ax = plt.axes()
sns.boxplot(x='DOL', y='Escherichia_Shigella', hue='DOL', data=df_concat, ax=ax, palette=colors, width=0.5, sym='', order=dol_order, dodge=False)

# Add swarm plot on top of boxplot inside each box group
sns.swarmplot(x='DOL', y='Escherichia_Shigella', hue='DOL', data=df_concat, palette=dot_colors, dodge=False, order=dol_order, hue_order=dol_order,size=7,legend=False)

# Set labels and title
plt.title('Escherichia_Shigella',fontsize=18)
plt.xlabel('DOL',fontsize=18)
plt.ylabel('Abundance',fontsize=18)

# Show the legend
plt.legend(title='DOL')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

# Actinobacteria
plt.figure(figsize=(10, 9))
ax = plt.axes()
sns.boxplot(x='DOL', y='Bacteroides', hue='DOL', data=df_concat, ax=ax, palette=colors, width=0.5, sym='', order=dol_order, dodge=False)

# Add swarm plot on top of boxplot inside each box group
sns.swarmplot(x='DOL', y='Bacteroides', hue='DOL', data=df_concat, palette=dot_colors, dodge=False, order=dol_order, hue_order=dol_order,size=7,legend=False)

# Set labels and title
plt.title('Bacteroides',fontsize=18)
plt.xlabel('DOL',fontsize=18)
plt.ylabel('Abundance',fontsize=18)

# Show the legend
plt.legend(title='DOL')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

# Show the plot
plt.show()
```
