```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Patch

# Define the data for the four pie charts
data_sets = [
    [11, 13, 25, 30],
    [48, 53, 113, 124],
    [59, 66, 138, 154],
    [59, 66, 138, 154]
]

# Define the colors for the pie charts
colors_sets = [
    ['#17becf', '#17becf', '#ff7f0e', '#ff7f0e'],
    ['#17becf', '#17becf', '#ff7f0e', '#ff7f0e'],
    ['#17becf', '#17becf', '#ff7f0e', '#ff7f0e'],
    ['#17becf', '#17becf', '#ff7f0e', '#ff7f0e']
]

# Define titles for each pie chart
titles = ['[1,2 | 1,2]', '[3 | 3]', '[1,2,3 | 1,2,3]', '[1,2,3 | 3]']

# Define hatch patterns and their descriptions
hatch_patterns = ['/', '|', '/', '|']  # Hatch patterns
hatch_descriptions = ['Pattern 1', 'Pattern 2', 'Pattern 3', 'Pattern 4']  # Descriptions for the legend
hatch_colors = ['black', 'red', 'green', 'blue']  # Colors for the hatch patterns

# Define the figure
fig, axs = plt.subplots(2, 2, figsize=(14, 12))
axs = axs.flatten()  # Flatten to 1D array for easier iteration

# Loop to create each pie chart with titles and updated colors
for idx, ax in enumerate(axs):
    sizes = data_sets[idx]
    explode = (0.05, 0.05, 0.01, 0.01)  # Adjust the explode values
    colors = colors_sets[idx]
    wedges, texts = ax.pie(sizes, colors=colors, startangle=90, explode=explode)

    # Adding hatch patterns and setting hatch colors
    for wedge, hatch, hatch_color in zip(wedges, hatch_patterns, hatch_colors):
        wedge.set_hatch(hatch)
        wedge.set_edgecolor(hatch_color)

    ax.set_title(titles[idx], fontsize=24)
    ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Define the dataset legend handles
dataset_handles = [
    Patch(facecolor=train_color, label='Train dataset'),
    Patch(facecolor=test_color, label='Test dataset')
]

# Define the hatch legend handles
hatch_handles = [Patch(facecolor='white', edgecolor=hatch_colors[i], hatch=hatch_patterns[i],
                       label=hatch_descriptions[i]) for i in range(len(hatch_patterns))]

# Combine all handles for the legend
all_handles = dataset_handles + hatch_handles

# Add the combined legend to the figure
fig.legend(handles=all_handles, loc='lower center', ncol=4, fontsize=12)

plt.tight_layout()
plt.show()
```
