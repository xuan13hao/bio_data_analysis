```python
import pandas as pd
import matplotlib.pyplot as plt

# Define a function to read a GTF file and count exon lengths
def read_gtf(gtf_file):
    exon_length_counts = {}
    with open(gtf_file, 'r') as file:
        for line in file:
            if line.startswith('#'):
                continue
            fields = line.strip().split('\t')
            if fields[2] == 'exon':
                start = int(fields[3])
                end = int(fields[4])
                length = end - start + 1
                if length in exon_length_counts:
                    exon_length_counts[length] += 1
                else:
                    exon_length_counts[length] = 1
    return exon_length_counts

# Replace this path with the path to your GTF file
gtf_file_path = 'Homo_sapiens.GRCh38.94.gtf'

# Read the GTF file and get the exon length counts
exon_length_counts = read_gtf(gtf_file_path)

# Create a DataFrame for the exon length counts
exon_length_df = pd.DataFrame(list(exon_length_counts.items()), columns=['Exon Length', 'Count'])

# Define ranges for exon lengths
bins = [0, 10, 50, 100, 500, float('inf')]
labels = ['1-10', '11-50', '51-100', '101-500', '>500']

# Create a new column for length range
exon_length_df['Length Range'] = pd.cut(exon_length_df['Exon Length'], bins=bins, labels=labels)

# Group by length range and count the occurrences
exon_length_range_counts = exon_length_df.groupby('Length Range')['Count'].sum().reset_index()
# exon_length_range_counts
# Plot the distribution of exon lengths by range
plt.figure(figsize=(10, 6))
bars = plt.bar(exon_length_range_counts['Length Range'], exon_length_range_counts['Count'], color='blue', edgecolor='black')
plt.title('Homo_sapiens.GRCh38 Distribution of Exon Lengths')
plt.xlabel('Exon Length')
plt.ylabel('Count')
plt.grid(False)

# Add count labels on top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2 - 0.1, yval + 50, int(yval), ha='left', va='bottom')

plt.show()
```
