```python
truth = pd.read_csv('sirv.csv',encoding = "ISO-8859-1")
truth
```

```python
d_exon = {}
d_aln = {}
for i in truth.index:
    zoom_truth = pd.Interval(truth.loc[i, 'start'],truth.loc[i, 'end'],closed='both')
    d_exon[i] = 0
    d_aln[i] = 0
    for j in df.index:
        zoom_df = pd.Interval(df.loc[j, 'sstart '],df.loc[j, 'send '],closed='both')
        if str(truth.loc[i, 'seqname']) == str(df.loc[j, 'sseqid ']) and zoom_truth.overlaps(zoom_df) and truth.loc[i, 'start'] <= df.loc[j, 'sstart '] and df.loc[j, 'send '] <= truth.loc[i, 'end'] :
            d_exon[i] = d_exon[i] +1
        if str(truth.loc[i, 'seqname']) == str(df.loc[j, 'sseqid ']) and zoom_truth.overlaps(zoom_df) :
            d_aln[i] = d_aln[i] +1
            
```

```python
exon_count_ = 0
for key, val in d_exon.items():
    if val > 0:
        exon_count_ = exon_count_ + 1
print("exon_count_ = ", exon_count_)
aln_count_ = 0
for key, val in d_aln.items():
    if val > 0:
        aln_count_ = aln_count_ + 1
print("aln_count_ = ", aln_count_)
```

```python
def bam_to_df(sam):
    ref = []
    seq = []
    name = []
    pos = []
    tlen = []
    end = []
    for read in sam:
        ref.append(read.rname)
        seq.append(len(read.seq))
        name.append(read.qname)
        pos.append(read.pos)
        tlen.append(len(read.cigar))
    return pd.DataFrame({'ref': ref,'qname':name ,'pos': pos, 'len': seq})

import pysam
from pandas import DataFrame
#myfile = pysam.AlignmentFile('aln_SIRV.sam', 'rb') 
#bam_to_df(myfile)
from simplesam import Reader, Writer
in_file = open('aln_SIRV.sam', 'r')
in_sam = Reader(in_file)
sam  = bam_to_df(in_sam)
```
