```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv('piechart_abundances_genus_C.csv')

# Sort the DataFrame by abundance in descending order
df_sorted = df.sort_values(by='Abundance', ascending=False)

# Extract top five taxa and their abundance values
top_five_taxa = df_sorted['Taxa'].iloc[:3].tolist()
top_five_abundance = df_sorted['Abundance'].iloc[:3].tolist()
other_abundance = df_sorted['Abundance'].iloc[3:].sum()

# Add 'Other' category
top_five_taxa.append('Other')
top_five_abundance.append(other_abundance)

# Define colors for each tax
colors = []
for tax in top_five_taxa:
    if tax == 'Lactobacillus':
        colors.append('#FF6347')
    elif tax == 'Escherichia_Shigella':
        colors.append('#4682B4')
    elif tax == 'Streptococcus':
        colors.append('#EE82EE')
    elif tax == 'Muribacter':
        colors.append('#F4A460')
    else:
        colors.append('#9ACD32')


# Plot a pie chart
plt.figure(figsize=(15, 8))
wedges, texts, autotexts = plt.pie(top_five_abundance, labels=None, autopct='%1.1f%%', startangle=140, colors=colors,pctdistance=1.1, labeldistance=1.6,shadow=True)
plt.title('')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
# Increase font size
for autotext in autotexts:
    autotext.set_fontsize(14)  # Change the font size as needed
# Add percentage labels on the pie chart
for autotext in autotexts:
    autotext.set_visible(True)

# Add legend
plt.legend(wedges, [f'{label}' for label, percentage in zip(top_five_taxa, top_five_abundance)],
           title='', loc='upper right', bbox_to_anchor=(1.1, 1), fontsize='large')


# Read the CSV file into a DataFrame
df = pd.read_csv('piechart_abundances_genus_F.csv')

# Sort the DataFrame by abundance in descending order
df_sorted = df.sort_values(by='Abundance', ascending=False)

# Extract top five taxa and their abundance values
top_five_taxa = df_sorted['Taxa'].iloc[:3].tolist()
top_five_abundance = df_sorted['Abundance'].iloc[:3].tolist()
other_abundance = df_sorted['Abundance'].iloc[3:].sum()

# Add 'Other' category
top_five_taxa.append('Other')
top_five_abundance.append(other_abundance)

# Define colors for each tax
colors = []
for tax in top_five_taxa:
    if tax == 'Lactobacillus':
        colors.append('#FF6347')
    elif tax == 'Escherichia_Shigella':
        colors.append('#4682B4')
    elif tax == 'Not_Assigned':
        colors.append('#FFE4C4')
    elif tax == 'Muribacter':
        colors.append('#F4A460')
    else:
        colors.append('#9ACD32')


# Plot a pie chart
plt.figure(figsize=(15, 8))
wedges, texts, autotexts = plt.pie(top_five_abundance, labels=None, autopct='%1.1f%%', startangle=140, colors=colors,pctdistance=1.1, labeldistance=1.6,shadow=True)
plt.title('')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
for autotext in autotexts:
    autotext.set_fontsize(14)  # Change the font size as needed
# Add percentage labels on the pie chart
for autotext in autotexts:
    autotext.set_visible(True)

# Add legend
plt.legend(wedges, [f'{label}' for label, percentage in zip(top_five_taxa, top_five_abundance)],
           title='', loc='upper right', bbox_to_anchor=(1.1, 1), fontsize='large')

plt.show()
```
