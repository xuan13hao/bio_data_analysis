```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Read data from CSV files for feeding and control
df_F = pd.read_csv('taxa_abund_Colon_het_p11-28.csv', index_col=0).T
df_F
df_F['Firmicutes'] = pd.to_numeric(df_F['Firmicutes'], errors='coerce')
df_F['Bacteroidetes'] = pd.to_numeric(df_F['Bacteroidetes'], errors='coerce')
df_F['Proteobacteria'] = pd.to_numeric(df_F['Proteobacteria'], errors='coerce')
df_concat = df_F

# Actinobacteria
plt.figure(figsize=(12, 10))
ax = plt.axes()
# Set specific colors for DOL
# Set specific colors for DOL
colors = {
    'P11': '#F8766D',  
          'P14': '#00BA38',
          'P28': '#619CFF'}
dot_colors = {
    'P11': 'black',  
          'P14': 'black',
          'P28': 'black'}
dol_order = ['P11', 'P14', 'P28']
# sns.boxplot(x='DOL', y='Lactobacillus', hue='DOL', data=df_F, ax=ax, palette=colors, width=0.8, sym='', order=dol_order)
# # Add swarm plot on top of boxplot inside each box group
# sns.swarmplot(x='DOL', y='Lactobacillus', hue='DOL', data=df_F, palette=dot_colors, dodge=True, order=dol_order)
# Plot data from df_concat
sns.boxplot(x='DOL', y='Firmicutes', hue='DOL', data=df_F, ax=ax, palette=colors, width=0.5, sym='', order=dol_order, dodge=False)

# Add swarm plot on top of boxplot inside each box group
sns.swarmplot(x='DOL', y='Firmicutes', hue='DOL', data=df_F, palette=dot_colors, dodge=False, order=dol_order, hue_order=dol_order,size=5,legend=False)

# Set labels and title
plt.title('Firmicutes',fontsize=18)
plt.xlabel('DOL',fontsize=18)
plt.ylabel('Abundance',fontsize=18)

# Show the legend
plt.legend(title='DOL')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

# Bacteroidetes
plt.figure(figsize=(12, 10))
ax = plt.axes()
# Set specific colors for DOL
# Set specific colors for DOL
colors = {
    'P11': '#F8766D',  
          'P14': '#00BA38',
          'P28': '#619CFF'}
dot_colors = {
    'P11': 'black',  
          'P14': 'black',
          'P28': 'black'}
dol_order = ['P11', 'P14', 'P28']
# sns.boxplot(x='DOL', y='Lactobacillus', hue='DOL', data=df_F, ax=ax, palette=colors, width=0.8, sym='', order=dol_order)
# # Add swarm plot on top of boxplot inside each box group
# sns.swarmplot(x='DOL', y='Lactobacillus', hue='DOL', data=df_F, palette=dot_colors, dodge=True, order=dol_order)
# Plot data from df_concat
sns.boxplot(x='DOL', y='Bacteroidetes', hue='DOL', data=df_F, ax=ax, palette=colors, width=0.5, sym='', order=dol_order, dodge=False)

# Add swarm plot on top of boxplot inside each box group
sns.swarmplot(x='DOL', y='Bacteroidetes', hue='DOL', data=df_F, palette=dot_colors, dodge=False, order=dol_order, hue_order=dol_order,size=5,legend=False)

# Set labels and title
plt.title('Bacteroidetes',fontsize=18)
plt.xlabel('DOL',fontsize=18)
plt.ylabel('Abundance',fontsize=18)

# Show the legend
plt.legend(title='DOL')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
# Show the plot

# Proteobacteria
plt.figure(figsize=(12, 10))
ax = plt.axes()
# Set specific colors for DOL
# Set specific colors for DOL
colors = {
    'P11': '#F8766D',  
          'P14': '#00BA38',
          'P28': '#619CFF'}
dot_colors = {
    'P11': 'black',  
          'P14': 'black',
          'P28': 'black'}
dol_order = ['P11', 'P14', 'P28']
# sns.boxplot(x='DOL', y='Lactobacillus', hue='DOL', data=df_F, ax=ax, palette=colors, width=0.8, sym='', order=dol_order)
# # Add swarm plot on top of boxplot inside each box group
# sns.swarmplot(x='DOL', y='Lactobacillus', hue='DOL', data=df_F, palette=dot_colors, dodge=True, order=dol_order)
# Plot data from df_concat
sns.boxplot(x='DOL', y='Proteobacteria', hue='DOL', data=df_F, ax=ax, palette=colors, width=0.5, sym='', order=dol_order, dodge=False)

# Add swarm plot on top of boxplot inside each box group
sns.swarmplot(x='DOL', y='Proteobacteria', hue='DOL', data=df_F, palette=dot_colors, dodge=False, order=dol_order, hue_order=dol_order,size=5,legend=False)

# Set labels and title
plt.title('Proteobacteria',fontsize=18)
plt.xlabel('DOL',fontsize=18)
plt.ylabel('Abundance',fontsize=18)

# Show the legend
plt.legend(title='DOL')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.show()
```
