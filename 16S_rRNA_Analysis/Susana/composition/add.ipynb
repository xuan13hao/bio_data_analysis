```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the CSV file
file_path = 'taxa_abund_phylum.csv'
taxa_abund_phylum = pd.read_csv(file_path)

# Extract the taxa names and groups
taxa_names = taxa_abund_phylum['Unnamed: 0']
groups = taxa_abund_phylum.loc[taxa_names == 'Response'].iloc[0, 1:].values

# Filter out the "Response" row
taxa_abund_phylum = taxa_abund_phylum[taxa_names != 'Response']

# Remove non-numeric columns
numeric_data = taxa_abund_phylum.iloc[:, 1:].apply(pd.to_numeric, errors='coerce')

# Convert abundances to relative abundances (percentages)
relative_abundances = numeric_data.div(numeric_data.sum(axis=0), axis=1) * 100

# Prepare the plot
fig, ax = plt.subplots(figsize=(12, 8))

# Define the color palette for the plot
colors = plt.get_cmap('tab20').colors

# Plot the stacked bar plot
bottom = None
for i, row in relative_abundances.iterrows():
    if bottom is None:
        bottom = row
        ax.bar(groups, row, label=taxa_names.iloc[i], color=colors[i % len(colors)])
    else:
        ax.bar(groups, row, bottom=bottom, label=taxa_names.iloc[i], color=colors[i % len(colors)])
        bottom += row

# Customize the plot
ax.set_ylabel('Relative abundance (%)')
ax.set_title('Relative Abundance of Bacterial Phyla in Different Groups')
ax.legend(loc='upper right', bbox_to_anchor=(1.2, 1), title='Phylum')

plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```
