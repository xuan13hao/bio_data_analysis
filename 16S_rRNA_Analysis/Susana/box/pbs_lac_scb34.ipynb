```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Read data from CSV file
df_F = pd.read_csv('taxa_abund_pbs_lac_scb34.csv', index_col=0).T

# Convert columns to numeric, coerce errors to NaN
df_F['Staphylococcus'] = pd.to_numeric(df_F['Staphylococcus'], errors='coerce')
df_F['Clostridium_sensu_stricto_1'] = pd.to_numeric(df_F['Clostridium_sensu_stricto_1'], errors='coerce')
df_F['Romboutsia'] = pd.to_numeric(df_F['Romboutsia'], errors='coerce')
df_F['Veillonella'] = pd.to_numeric(df_F['Veillonella'], errors='coerce')
df_F['Lactobacillus'] = pd.to_numeric(df_F['Lactobacillus'], errors='coerce')
df_F['Escherichia_Shigella'] = pd.to_numeric(df_F['Escherichia_Shigella'], errors='coerce')

# Define the list of features to plot
features = ['Staphylococcus', 'Clostridium_sensu_stricto_1', 'Romboutsia','Veillonella','Escherichia_Shigella','Lactobacillus']

# Plotting each feature in a loop
for feature in features:
    plt.figure(figsize=(8, 6))
    ax = plt.axes()
    
    # Set specific colors for DOL
    colors = {
        'PBS-SCB34': '#00BFC4',
        'Lac-SCB34': '#C77CFF'
    }
    dot_colors = {
        'PBS-SCB34': 'black',
        'Lac-SCB34': 'black'
    }
    dol_order = ['PBS-SCB34', 'Lac-SCB34']

    # Create boxplot
    sns.boxplot(x='Response', y=feature, hue='Response', data=df_F, ax=ax, palette=colors, width=0.6, order=dol_order, dodge=False, showmeans=True, meanline=True, showcaps=False, fliersize=0)

    # Add swarm plot on top of boxplot
    sns.swarmplot(x='Response', y=feature, hue='Response', data=df_F, palette=dot_colors, dodge=False, order=dol_order, size=5)

    # Set labels and title
    plt.title(f'{feature}', fontsize=18)
    plt.xlabel('Response', fontsize=18)
    plt.ylabel('Abundance', fontsize=18)

    # Remove the duplicate legend from swarmplot
    # handles, labels = ax.get_legend_handles_labels()
    # plt.legend(handles[:len(dol_order)], labels[:len(dol_order)], title='Response')

    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)

    # Display the plot
    plt.show()
```
