```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import re
```

```python
df = pd.read_csv("alphadiversity.csv")
```

```python
plt.figure(figsize=(10, 6))  # Set the figure size

# Assuming your CSV has columns 'Group' and 'Value'
# Use the boxplot method with 'x' as the grouping variable and 'y' as the values
plt.boxplot([group['value'] for name, group in df.groupby('Response')], labels=df['Response'].unique())

# You can customize the plot further, like adding labels, titles, etc.
plt.xlabel('Groups')
plt.ylabel('Values')
plt.title('Box Plot with Different Groups')

# Add data points as dots
for i, group_name in enumerate(df['Response'].unique()):
    y_values = df[df['Response'] == group_name]['value']
    x_values = [i + 1] * len(y_values)  # Position on the x-axis
    plt.scatter(x_values, y_values, label=group_name, alpha=0.7)

plt.legend()  # Show the legend with group names

plt.show()
```
